```
from colorama import Fore,Back,Style


statement_dictionary={
    'poperror':Back.LIGHTRED_EX + Fore.WHITE+Style.BRIGHT+"YOU CANNOT POP HERE  :(:(:(",
    'pusherror':Back.LIGHTRED_EX + Fore.WHITE+Style.BRIGHT+"YOU CANNOT PUSH HERE  :(:(:(",
    'success':Fore.BLACK+Back.LIGHTGREEN_EX+Style.BRIGHT+ "$$$$$$$$$$  CONGRATS YOU WON THE GAME  :):):):)$$$$$$$$$$"
     
}

#pop function
def pop_fun(t,l):
  #wrong bucket
  if(t<1 or t>3):
    print(statement_dictionary['poperror'])
    return -1
  else:
    #there is no color
    if(len(l[t-1])==0):
      print(statement_dictionary['poperror'])
      return -1
    else:
      k=l[t-1].pop()
      return k
    

#push function 
def push_fun(t1,k,l,t2):
  #wrong bucket
  if(t1<1 or t1>3):
    print(statement_dictionary['pusherror'])
    l[t2-1].append(k)
    return l
  elif(len(l[t1-1])==0):
    l[t1-1].append(k)
    return l
    #different colors
  elif(l[t1-1][-1]!=k or k==-1):
    print(statement_dictionary['pusherror'])
    l[t2-1].append(k)
    return l
    #four colors already present
  elif(len(l[t1-1])==4):
    print(statement_dictionary['pusherror'])
    l[t2-1].append(k)
    return l
  else:
    l[t1-1].append(k)
    return l


#move function
def move(l):
  print(Fore.LIGHTBLUE_EX+Back.RESET+Style.BRIGHT+"ENTER THE BUCKET NUMBER FROM WHERE YOU WANT TO POP ")
  t1=int(input())
  k=pop_fun(t1,l)
  if(k==-1):
    return l
  print(Fore.LIGHTBLUE_EX+Back.RESET+Style.BRIGHT+ "ENTER THE BUCKET NUMBER TO WHERE YOU WANT TO PUSH ")
  t2=int(input())
  return push_fun(t2,k,l,t1)

#print function
def print_data(l):
  print()
  print()
  for i in range(len(l)):
    print(Fore.BLACK+Back.RESET+Style.BRIGHT+ "bucket {}".format(i+1))
    print()
    for j in range(len(l[i])):
      print(l[i][len(l[i])-j-1])
    print()
  print()
  print()

#check_win
def check_win(l):
  c=0
  for i in range(len(l)):
    if(len(l[i])!=0):
      if(l[i].count(l[i][0])==4):
        c+=1
  if(c==2):
    return 1
  else:
    return 0
 
  
#code starts

print(Back.RESET+ Fore.LIGHTMAGENTA_EX+"*****************************COLOR SORTING********************************")
print()
print()
print(Back.RESET+ Fore.LIGHTRED_EX+"READ THE BELOW INSTRUCTIONS")
print()
print(Back.RESET+ Fore.LIGHTBLACK_EX+"1->YOU CAN ONLY MOVE ONE COLOR FROM A BUCKET WHICH IS AT THE TOP")
print(Back.RESET+ Fore.LIGHTBLACK_EX+"2->YOU CANNOT PLACE A COLOR OVER TOP OF DIFFERENT COLOUR")
print(Back.RESET+ Fore.LIGHTBLACK_EX+"3->EACH BUCKET CAN HOLD AT MOST 4 COLOURS")
print()
print()
print(Back.RESET+ Fore.GREEN+"**********************************LETS BEGIN THE GAME*****************************")
        
    





green=Back.LIGHTGREEN_EX+ Fore.LIGHTGREEN_EX+ "texttext"
yellow=Back.LIGHTYELLOW_EX + Fore.LIGHTYELLOW_EX +"texttext"
l=[[green,yellow,green,yellow],[yellow,green,yellow],[green]]



f=0
print_data(l)

while(f==0):
  l=move(l)
  print_data(l)
  f=check_win(l)
  if(f==1):
    print(statement_dictionary['success'])








  
```
